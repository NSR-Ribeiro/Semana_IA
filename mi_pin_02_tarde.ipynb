```python
from sklearn.model_selection import train_test_split
```

```python
treino, teste, classe_treino, classe_teste = train_test_split(df_resenhas.text_pt, df_resenhas.sentiment,random_state=42)
```

# Escolhendo  slgoritmo de aprendizado supervisionado - Classificação

```python
treino
```

```python
teste
```

```python
classe_treino
```

```python
classe_teste
```

```python
from sklearn.linear_model import LogisticRegression

# criei um objeto   | Chamada do construtor da classe logisticRegression

regressao_logistica = LogisticRegression()

# treinamento

regressao_logistica.fit(treino, classe_treino) # rótulos - neg pos

# acurácia

acuracia = regressao_logistica.score(teste, classe_teste)

print(acuracia)
```

# Exercicio 02 - vamos mudar o neg e pos para 0 e 1?

```python
# Criei uma variável classificação.
# Acessei a coluna sentimento
# Usei  a função replace para trocar "neg pos" por" "0 e 1"

classificacao = df_resenhas.sentiment.replace(["neg", "pos"], [0,1])
```
