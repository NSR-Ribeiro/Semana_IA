# API Pokemons

```python
# Importando bibliotecas
import requests

# URL da API

url = "https://pokeapi.co/api/v2/pokemon/1"

# Enviando a requisição

response = requests.get(url)

print(response.text)

if response.status_code == 200:
  data = response.json()

  print(f"Nome: {data['name']}") #Removed extra indent
  print(f"Altura: {data['id']}") #Removed extra indent
  print(f"ID: {data['weight']}") #Removed extra indent
  print(f"Peso: {data['height']}") #Removed extra indent

  print("tipos:") #Removed extra indent

  for tipo in data['types']: #Removed extra indent
      print(f"- {tipo['type']['name']}")
else:
  print("Erro na requisição")
```

# API star wars

```python
# Definindo endoint
endpoint = "/people"

# Definindo URL da API
url = f"https://swapi.dev/api{endpoint}"

# Fazendo a requisição
response = requests.get(url).json()

# Imprimindo os resultados
print(response)

# Iterando no dicionario retornando
for pessoa in response['results']:
    print("="*40)
    print(f"Nome: {pessoa['name']}")
    print(f"Género: {pessoa['gender']}")
    print(f"Altura: {pessoa['height']} cm")
    print(f"Peso: {pessoa['mass']} kg")
    print("Participação em:")

# Iterando em cada filme

    for filme_url in pessoa['films']:
      filme_response = requests.get(filme_url).json()
      print(f"- {filme_response['title']}")






```
