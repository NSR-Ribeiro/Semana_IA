```python
import requests
from bs4 import BeautifulSoup

# URL do site
url = "https://www.gft.com/br/pt/technology/artificial-intelligence"

# Fazer a requisição para obter o conteúdo da página
response = requests.get(url)
html_content = response.text

# Analisar o conteúdo HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Extrair o texto da página
text = soup.get_text(separator=' ')
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Baixar recursos necessários do nltk
nltk.download('punkt')
nltk.download('stopwords')

# Tokenização
tokens = word_tokenize(text)

# Remover pontuações e converter para minúsculas
words = [word.lower() for word in tokens if word.isalnum()]

# Remover stopwords
stop_words = set(stopwords.words('portuguese'))
words = [word for word in words if word not in stop_words]

# Calcular a frequência das palavras
freq_dist = FreqDist(words)

# Plotar a distribuição de frequência
plt.figure(figsize=(12, 6))
freq_dist.plot(30, cumulative=False)
plt.show()

```

```python
# Gerar e mostrar a nuvem de palavras
wordcloud = WordCloud(width=15000, height=7000, background_color='white').generate(' '.join(words))

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Não mostrar eixos
plt.show()
```
